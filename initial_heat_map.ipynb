```python
# Import libraries and required dependencies
import pandas as pd
import numpy as np
import seaborn as sns
from pathlib import Path
import matplotlib
sns.set_theme(style="whitegrid")
```

```python
# Create path to csv for reading
df_initial_kc_csv = Path('./csv_to_read/KCWT_csv_file.csv')
df_initial_kcg_csv = Path('./csv_to_read/KCG_csv_file.csv')
# kc_csv = Path('../<INSERT FILE PATH HERE>')
# kcg_csv = Path('../<INSERT FILE PATH HERE>')
```

```python
# Read csv to dataframe
df_initial_kc = pd.read_csv(df_initial_kc_csv, index_col='section')
df_initial_kcg = pd.read_csv(df_initial_kcg_csv, index_col='section')
# Review dataframe
df_initial_kcg 
```

```python
# replace no values with 0
df_initial_kc = df_initial_kc.fillna(0)
df_initial_kc
```

```python
# change objects to floats
df_initial_kc.dtypes
df_initial_kc['kc1_pct_legion'] = df_initial_kc['kc1_pct_legion']/100
```

```python
g = sns.relplot(
    data=df_initial_kc,
    x="section", 
    y="kc1_score",  
    
    size="kc1_pct_legion",
    palette="vlag", 
    hue_norm=(-1, 1), 
    edgecolor=".7",
    height=10, 
    sizes=(50, 250), 
    size_norm=(-.2, .8),
)

# Tweak the figure to finalize
g.set(xlabel="", ylabel="", aspect="equal")
g.despine(left=True, bottom=True)
g.ax.margins(.02)
for label in g.ax.get_xticklabels():
    label.set_rotation(90)
for artist in g.legend.legendHandles:
    artist.set_edgecolor(".7")
```
