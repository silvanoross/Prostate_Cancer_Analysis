```python
# import necessary libraries and dependencies

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from pathlib import Path
```

```python
# Read csv kcwt
kcg_csv = Path('./KCG_initial.csv')
```

```python
# Read into pd dataframe
kcg_df = pd.read_csv(kcg_csv)
kcg_df.head()
```

```python
# Replace ADM score with 1
kcg_df.loc[:, ['KCG_adm', 'KCG_adm.1', 'KCG_adm.2', 'KCG_adm.3']] = kcg_df.loc[:, ['KCG_adm', 'KCG_adm.1', 'KCG_adm.2', 'KCG_adm.3']].replace(['A'], 1).fillna(0)
kcg_df.head()
```

```python
# insert new column into dataframe
kcg_mouse_id_df = kcg_df
kcg_mouse_id_df.insert(1, 'id_1', 1)
```

```python
# insert mouse ID 2
kcg_mouse_id_df.insert(5, 'id_2', 2)
kcg_mouse_id_df.head()
```

```python
# insert mouse ID 3
kcg_mouse_id_df.insert(9, 'id_3', 3)
kcg_mouse_id_df.head()
```

```python
# insert mouse ID 4
kcg_mouse_id_df.insert(13, 'id_4', 4)
kcg_mouse_id_df.head()
```

```python
kcg_mouse_id_df['id_1'] = 1 
kcg_mouse_id_df['id_2'] = 2
kcg_mouse_id_df['id_3'] = 3
kcg_mouse_id_df['id_4'] = 4
kcg_mouse_id_df.head()
```

```python
# view head of dataframe for KCG mice
kcg_mouse_id_df.head()
```

```python
# use the scatterplot function to build the bubble map
# kc1
plot_1 = sns.scatterplot(data=kcg_mouse_id_df, x="Unnamed: 0", y="id_1", 
                         size="KCG_score", legend=False, sizes=(5, 200), 
                         hue='KCG_adm')

# show the graph
plt.show(plot_1)
```

```python
# restructure data
kcg_1_df = (kcg_mouse_id_df.loc[:, ['Unnamed: 0', 'id_1', 'KCG_score', 'KCG_adm']].assign(group = 'id_1').set_axis(['field', 'id', 'score', 'adm', 'id_1'], axis=1))

kcg_2_df = (kcg_mouse_id_df.loc[:, ['Unnamed: 0', 'id_2', 'KCG_score.1', 'KCG_adm.1']].assign(group = 'id_2').set_axis(['field', 'id', 'score', 'adm', 'id_2',], axis=1))

kcg_3_df = (kcg_mouse_id_df.loc[:, ['Unnamed: 0', 'id_3', 'KCG_score.2', 'KCG_adm.2']].assign(group = 'id_3').set_axis(['field', 'id', 'score', 'adm', 'id_3'], axis=1))


kcg_4_df = (kcg_mouse_id_df.loc[:, ['Unnamed: 0', 'id_4', 'KCG_score.3', 'KCG_adm.3']].assign(group = 'id_4').set_axis(['field', 'id', 'score', 'adm', 'id_4'], axis=1))
```

```python
kcg_1_df
```

```python
# merge dataframes
merged_dataframe = pd.concat([kcg_1_df, kcg_2_df, kcg_3_df, kcg_4_df], ignore_index=True)
```

```python
# plot merged df
# configure plot
sns.set_style("whitegrid")
cmap = ['gray', 'blue']


g = sns.relplot(data=merged_dataframe, 
            x='field', 
            y='id',
            hue='adm', 
            size='score',
            sizes=(50, 500),
            height=6,
            aspect=2.5,
            palette=cmap,
            )
# Change axis titles
g.set(title="4 KCG MICE", xlabel='Fields', ylabel='Mouse')

#Export Graph
plt.savefig('KCG_grid_plot.pdf', dpi=300)
```
