```python
# import necessary libraries and dependencies

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from pathlib import Path
```

```python
# Read csv kcwt
kc_csv = Path('./KCWT.csv')
```

```python
# Read into pd dataframe
kc_df = pd.read_csv(kc_csv)
kc_df.head()
```

```python
# Replace ADM score with 1
kc_df.loc[:, ['KC_adm', 'KC_adm.1', 'KC_adm.2', 'KC_adm.3']] = kc_df.loc[:, ['KC_adm', 'KC_adm.1', 'KC_adm.2', 'KC_adm.3']].replace(['A'], 1)
kc_df.head()
```

```python
# insert new column into dataframe
kc_mouse_id_df = kc_df
kc_mouse_id_df.insert(1, 'id_1', 1)
```

```python
# insert mouse ID 2
kc_mouse_id_df.insert(5, 'id_2', 2)
kc_mouse_id_df.head()
```

```python
# insert mouse ID 3
kc_mouse_id_df.insert(9, 'id_3', 3)
kc_mouse_id_df.head()
```

```python
# insert mouse ID 4
kc_mouse_id_df.insert(13, 'id_4', 4)
kc_mouse_id_df.head()
```

```python
kc_mouse_id_df['id_1'] = 1 
kc_mouse_id_df['id_2'] = 1.02
kc_mouse_id_df['id_3'] = 1.04
kc_mouse_id_df['id_4'] = 1.06
kc_mouse_id_df.head()
```

```python
kc_mouse_id_df = kc_mouse_id_df.fillna(0)
kc_mouse_id_df.head()
```

```python
ax = plt.subplots()
```

```python
# use the scatterplot function to build the bubble map
# kc1
plot_1 = sns.scatterplot(data=kc_mouse_id_df, x="Unnamed: 0", y="id_1", 
                         size="KC_score", legend=False, sizes=(5, 200), 
                         hue='KC_adm')

# show the graph
plt.show(plot_1)
```

```python
# create kc2
ax = sns.scatterplot(legend=False,
                     data=kc_mouse_id_df, x="Unnamed: 0", y="id_2", 
                     size="KC_score.1", sizes=(5, 200), 
                     hue='KC_adm.1')
```

```python
# create kc3
plot_3 = sns.scatterplot(data=kc_mouse_id_df, x="Unnamed: 0", y="id_3", 
                         size="KC_score.2", legend=False, sizes=(5, 200), 
                         hue='KC_adm.2')
```

```python
# create kc2
plot_4 = sns.scatterplot(data=kc_mouse_id_df, x="Unnamed: 0", y="id_4", 
                         size="KC_score.3", legend=False, sizes=(5, 200), 
                         hue='KC_adm.3')
```

```python
# create overlays
plt.figure(figsize=(6,15))
```

```python
plot_replot_1 = sns.relplot(data=kc_mouse_id_df, x="Unnamed: 0", y="id_1", 
                         size="KC_score", legend=False, sizes=(5, 200), 
                         hue='KC_adm')

plot_replot_2 = sns.kdeplot(data=kc_mouse_id_df, x="Unnamed: 0", y="id_2", 
                         size="KC_score.1", legend=False, sizes=(5, 200), 
                         hue='KC_adm.1')
```

```python
# restructure data
kc_1_df = (kc_mouse_id_df.loc[:, ['Unnamed: 0', 'id_1', 'KC_score']].assign(group = 'id_1').set_axis(['field', 'id', 'score', 'id_1'], axis=1))

kc_2_df = (kc_mouse_id_df.loc[:, ['Unnamed: 0', 'id_2', 'KC_score.1']].assign(group = 'id_2').set_axis(['field', 'id', 'score', 'id_2'], axis=1))

kc_3_df = (kc_mouse_id_df.loc[:, ['Unnamed: 0', 'id_3', 'KC_score.2']].assign(group = 'id_3').set_axis(['field', 'id', 'score', 'id_3'], axis=1))


kc_4_df = (kc_mouse_id_df.loc[:, ['Unnamed: 0', 'id_4', 'KC_score.3']].assign(group = 'id_4').set_axis(['field', 'id', 'score', 'id_4'], axis=1))
```

```python
# merge dataframes
merged_dataframe = pd.concat([kc_1_df, kc_2_df, kc_3_df, kc_4_df], ignore_index=True)
```

```python
# plot merged df
sns.set_style("whitegrid")

sns.relplot(data=merged_dataframe, x='field', y='id', hue='score', size='score')
```
