```python
# import necessary libraries and dependencies

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from pathlib import Path
```

```python
# Read csv kcwt
kc_csv = Path('./KCWT_initial.csv')
```

```python
# Read into pd dataframe
kc_df = pd.read_csv(kc_csv)
kc_df.head()
```

```python
# Replace ADM score with 1
kc_df.loc[:, ['KC_adm', 'KC_adm.1', 'KC_adm.2', 'KC_adm.3']] = kc_df.loc[:, ['KC_adm', 'KC_adm.1', 'KC_adm.2', 'KC_adm.3']].replace(['A'], 1).fillna(0)
kc_df
```

```python
# insert new column into dataframe
kc_mouse_id_df = kc_df
kc_mouse_id_df.insert(1, 'id_1', 1)
```

```python
# insert mouse ID 2
kc_mouse_id_df.insert(5, 'id_2', 2)
kc_mouse_id_df.head()
```

```python
# insert mouse ID 3
kc_mouse_id_df.insert(9, 'id_3', 3)
kc_mouse_id_df.head()
```

```python
# insert mouse ID 4
kc_mouse_id_df.insert(13, 'id_4', 4)
kc_mouse_id_df.head()
```

```python
kc_mouse_id_df['id_1'] = 1 
kc_mouse_id_df['id_2'] = 2
kc_mouse_id_df['id_3'] = 3
kc_mouse_id_df['id_4'] = 4
kc_mouse_id_df.head()
```

```python

kc_mouse_id_df.head()
```

```python
# use the scatterplot function to build the bubble map
# kc1
plot_1 = sns.scatterplot(data=kc_mouse_id_df, x="Unnamed: 0", y="id_1", 
                         size="KC_score", legend=False, sizes=(5, 200), 
                         hue='KC_adm')

# show the graph
plt.show(plot_1)
```

```python
# create kc2
ax = sns.scatterplot(legend=False,
                     data=kc_mouse_id_df, x="Unnamed: 0", y="id_2", 
                     size="KC_score.1", sizes=(5, 200), 
                     hue='KC_adm.1')
```

```python
# create kc3
plot_3 = sns.scatterplot(data=kc_mouse_id_df, x="Unnamed: 0", y="id_3", 
                         size="KC_score.2", legend=False, sizes=(5, 200), 
                         hue='KC_adm.2')
```

```python
# create kc2
plot_4 = sns.scatterplot(data=kc_mouse_id_df, x="Unnamed: 0", y="id_4", 
                         size="KC_score.3", legend=False, sizes=(5, 200), 
                         hue='KC_adm.3')
```

```python
# create overlays
plt.figure(figsize=(6,15))
plt.show()
```

```python
# restructure data
kc_1_df = (kc_mouse_id_df.loc[:, ['Unnamed: 0', 'id_1', 'KC_score', 'KC_adm']].assign(group = 'id_1').set_axis(['field', 'id', 'score', 'adm', 'id_1'], axis=1))

kc_2_df = (kc_mouse_id_df.loc[:, ['Unnamed: 0', 'id_2', 'KC_score.1', 'KC_adm.1']].assign(group = 'id_2').set_axis(['field', 'id', 'score', 'adm', 'id_2',], axis=1))

kc_3_df = (kc_mouse_id_df.loc[:, ['Unnamed: 0', 'id_3', 'KC_score.2', 'KC_adm.2']].assign(group = 'id_3').set_axis(['field', 'id', 'score', 'adm', 'id_3'], axis=1))


kc_4_df = (kc_mouse_id_df.loc[:, ['Unnamed: 0', 'id_4', 'KC_score.3', 'KC_adm.3']].assign(group = 'id_4').set_axis(['field', 'id', 'score', 'adm', 'id_4'], axis=1))
```

```python
# merge dataframes
merged_dataframe = pd.concat([kc_1_df, kc_2_df, kc_3_df, kc_4_df], ignore_index=True)
```

```python
# plot merged df
sns.set_style("whitegrid")
cmap = ['gray', 'blue']
g = sns.relplot(data=merged_dataframe, 
            x='field', 
            y='id', 
            hue='adm', 
            size='score',
            sizes=(50, 500),
            height=4,
            aspect=2,
            palette=cmap)

# Change axis titles
g.set(title="4 KC MICE", xlabel='Fields', ylabel='Mouse')

# Change font size
g.tight_layout(w_pad=0)

# Save fig
plt.savefig('KC_grid_plot.pdf', dpi=300)
```
